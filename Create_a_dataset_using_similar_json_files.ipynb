# Importing Libraries

```python
import numpy as np
import pandas as pd

import json 

import warnings
# Filter warnings
warnings.filterwarnings('ignore')
```

#Loading json file from google drive and saving it as dataset

```python
with open('/content/drive/MyDrive/sample1.json', 'r') as f:
    data1 = json.load(f)
df1 = (pd.DataFrame({'a': data1})).T
df1
```

```python
with open('/content/drive/MyDrive/sample2.json', 'r') as f:
    data2 = json.load(f)
df2 = (pd.DataFrame({'b': data2})).T
df2
```

```python
with open('/content/drive/MyDrive/sample3.json', 'r') as f:
    data3 = json.load(f)
df3 = (pd.DataFrame({'c': data3})).T
df3
```

```python
with open('/content/drive/MyDrive/sample4.json', 'r') as f:
    data4 = json.load(f)
df4 = (pd.DataFrame({'d': data4})).T
df4
```

#Combine all observation in different json files to a dataset

```python
df = pd.concat([df1,df2,df3,df4])
df
```
